```python
import matplotlib.pyplot as plt
import numpy as np

# Função para adicionar os valores no topo das barras
def autolabel(bars, ax):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# 1. Gráfico de Vendas por Região
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
vendas_2022 = np.array([1.2, 2.3, 1.5, 3.8, 1.7])  # em milhões
vendas_2023 = np.array([1.0, 2.1, 1.3, 3.2, 1.5])  # em milhões

fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = np.arange(len(regioes))

bar1 = ax.bar(index, vendas_2022, bar_width, label='Dez/2022')
bar2 = ax.bar(index + bar_width, vendas_2023, bar_width, label='Dez/2023')

ax.set_xlabel('Regiões')
ax.set_ylabel('Vendas (em milhões de R$)')
ax.set_title('Comparação de Vendas por Região - Dez/2022 vs Dez/2023')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(regioes)
ax.legend()

autolabel(bar1, ax)
autolabel(bar2, ax)

plt.tight_layout()
plt.show()

# 2. Gráfico de Vendas por Tipo de Cliente
clientes = ['Novos Clientes', 'Clientes Antigos']
vendas_2022_clientes = [2.5, 8.0]  # em milhões
vendas_2023_clientes = [2.0, 6.8]  # em milhões

fig, ax = plt.subplots(figsize=(8, 6))
index = np.arange(len(clientes))  # redefine o index para os tipos de clientes

bar1 = ax.bar(index, vendas_2022_clientes, bar_width, label='Dez/2022')
bar2 = ax.bar(index + bar_width, vendas_2023_clientes, bar_width, label='Dez/2023')

ax.set_xlabel('Tipo de Cliente')
ax.set_ylabel('Vendas (em milhões de R$)')
ax.set_title('Comparação de Vendas por Tipo de Cliente - Dez/2022 vs Dez/2023')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(clientes)
ax.legend()

autolabel(bar1, ax)
autolabel(bar2, ax)

plt.tight_layout()
plt.show()

# 3. Gráfico de Vendas por Categoria de Produto
categorias = ['Celulares', 'Desktops', 'Notebooks', 'Smartwatches', 'Acessórios']
vendas_2022_produtos = [8.0, 3.0, 5.0, 2.5, 1.5]  # em milhões
vendas_2023_produtos = [6.0, 2.8, 4.7, 2.3, 1.4]  # em milhões

fig, ax = plt.subplots(figsize=(10, 6))
index = np.arange(len(categorias))  # redefine o index para as categorias de produto

bar1 = ax.bar(index, vendas_2022_produtos, bar_width, label='Dez/2022')
bar2 = ax.bar(index + bar_width, vendas_2023_produtos, bar_width, label='Dez/2023')

ax.set_xlabel('Categoria de Produto')
ax.set_ylabel('Vendas (em milhões de R$)')
ax.set_title('Comparação de Vendas por Categoria de Produto - Dez/2022 vs Dez/2023')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(categorias)
ax.legend()

autolabel(bar1, ax)
autolabel(bar2, ax)

plt.tight_layout()
plt.show()

```
